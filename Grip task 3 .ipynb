### Data Visualization

```python
plt.figure(figsize  = (10,6))
sns.countplot(x='Ship Mode', data=df)
plt.show()
```

```python
plt.figure(figsize  = (10,6))
sns.countplot(x='Segment', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title('Count of State wise shipments')
plt.show()
```

### Visualization  Understanding  of Count of State Wise Shipments : 

##### State wise California has highest sales.

##### States like Texas, New York, etc are active in business.

##### Few states like North Dakota, Wyoming and West Virginia have less or no sales at all.

##### To generate more profit we have to increase our sales in Texas and New York where we have potential customers.

```python
plt.figure(figsize = (9,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

##### This pie graph indicates that the  Office supplies have  the highest demand and others are comparatively less in demand

```python
plt.figure(figsize = (10,6))
df.groupby('Region')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

### Overview  Understannding of Profit in different Region Chart  

##### West generates most of the profit followed by East.

##### Central and Southern region are generating less profit as compared to the Eastern and Western Region.

##### Increase the sales in Central and Southern region in order to generate more profit in these region.

```python
sp = df.groupby('State')['Profit'].sum()
plt.figure(figsize=(16,8))
sns.barplot(x=sp.index, y=sp.values)
plt.xticks(rotation=90)
plt.ylabel('Profit')
plt.show()
```

**From the above graph we can see california and new york has hishest profit and virginia,lowa,and some with low and zero profit**

```python
sns.pairplot(df, hue='Segment')
```

```python
plt.figure(figsize = (9,5))
sns.heatmap(df.corr(), annot = True, cmap = "PuBu")
```

```python
df.groupby(['Category'])[['Profit','Sales']].agg(sum).plot(kind='bar',figsize=(10,6), )
plt.ylabel('Profit and Sales')
plt.show()
```

##### This shows  that Furniture is generating very less profit.

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Category',hue='Segment',data=df)
plt.ylabel('Count of Segment')
plt.title('Count of Segment by Category and Segment')
```

##### The highest purchases are in consumer segment whereas home office segment has the least purchases in each category. We need to focus on the corporate and home office segment as well to increase our sales and profit.

```python
df.groupby('Sub-Category')[['Sales','Profit']].agg(sum).plot(kind='barh',figsize=(15, 8))
plt.ylabel('Sub-Category')
plt.xlabel('Sales and Profit')
plt.show()
```

**From the above graph phones and chair has a highest sales and copiers and phones has highest profit respectively.**`

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Region',hue='Ship Mode',data=df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

##### Least sales are in Southern region, we can attract this region by offering more discount. Standard mode of shipment is opted by most of the customers in each region.

### Weak Areas Where You Can Work To Make More Profit.

##### 3 After the highest sales in California we have high value of potential customers in Texas and New York so we should concentrate more on these states to generate more profit.

##### 4 We have highest purchases in consumer segment but we should also concentrate on corporate and home office segments to increase our sales and profit.
